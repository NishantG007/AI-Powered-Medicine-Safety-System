```python
import json
from collections import Counter
```

```python
FILE_PATH = "D:/Programming/AI_Medicine_Safety_Platform/medicine_data.json"

try:
    with open(FILE_PATH, "r", encoding="utf-8") as file:
        medicines_data = json.load(file)

    print("✅ JSON loaded successfully.\n")

except json.JSONDecodeError as e:
    print("❌ Broken JSON detected.")
    print(f"Error: {e}")
    exit()

except FileNotFoundError:
    print("❌ medicines.json file not found.")
    exit()
```

```python
required_fields = [
    "medicine_name",
    "category",
    "common_use",
    "side_effects",
    "severity_level",
    "precautions",
    "interactions"
]
```

#CHECK FOR MISSING FIELDS

```python
print("========== CHECKING MISSING FIELDS ==========\n")

missing_field_count = 0

for idx, med in enumerate(medicines_data):

    missing = []

    for field in required_fields:
        if field not in med:
            missing.append(field)

    if missing:
        missing_field_count += 1
        print(f"❌ Entry {idx + 1} missing fields: {missing}")

if missing_field_count == 0:
    print("✅ No missing fields found.\n")
```

#CHECK DUPLICATE MEDICINES

```python
print("\n========== CHECKING DUPLICATE MEDICINES ==========\n")

medicine_names = []

for med in medicines_data:

    if "medicine_name" in med:
        medicine_names.append(
            med["medicine_name"].strip().lower()
        )

name_counts = Counter(medicine_names)

duplicates = [
    name for name, count in name_counts.items()
    if count > 1
]

if duplicates:

    print("❌ Duplicate medicines found:\n")

    for dup in duplicates:
        print(f"- {dup}")

else:
    print("✅ No duplicate medicines found.\n")
```

#CHECK INCONSISTENT NAMES

```python
print("\n========== CHECKING INCONSISTENT NAMES ==========\n")

original_names = []

for med in medicines_data:

    if "medicine_name" in med:
        original_names.append(med["medicine_name"])

normalized_map = {}

for name in original_names:

    normalized = name.strip().lower()

    if normalized not in normalized_map:
        normalized_map[normalized] = []

    normalized_map[normalized].append(name)

inconsistencies_found = False

for normalized, versions in normalized_map.items():

    unique_versions = list(set(versions))

    if len(unique_versions) > 1:

        inconsistencies_found = True

        print(f"⚠ Inconsistent naming for '{normalized}':")
        print(unique_versions)
        print()

if not inconsistencies_found:
    print("✅ No inconsistent naming found.\n")
```

#CHECK INTERACTION CONSISTENCY

```python
print("\n========== CHECKING INTERACTION CONSISTENCY ==========\n")

# Build medicine lookup
medicine_lookup = {}

for med in medicines_data:

    name = med.get("medicine_name", "").strip()

    medicine_lookup[name.lower()] = med


inconsistency_count = 0

for med in medicines_data:

    med_name = med.get("medicine_name", "").strip()

    interactions = med.get("interactions", {})

    for interacting_med in interactions.keys():

        interacting_med_lower = interacting_med.strip().lower()

        # Check if interacting medicine exists
        if interacting_med_lower in medicine_lookup:

            interacting_entry = medicine_lookup[interacting_med_lower]

            reverse_interactions = interacting_entry.get(
                "interactions", {}
            )

            # Check reverse interaction
            if med_name not in reverse_interactions:

                inconsistency_count += 1

                print(
                    f"⚠ Missing reverse interaction:"
                )

                print(
                    f"{interacting_med} -> {med_name}"
                )

                print()

if inconsistency_count == 0:
    print("✅ All interactions are consistent.\n")
```

```python
print("\n========== DATASET VALIDATION COMPLETE ==========\n")

print(f"Total Medicines: {len(medicines_data)}")
print(f"Duplicate Medicines Found: {len(duplicates)}")
print(f"Missing Field Entries: {missing_field_count}")
print(f"Interaction Inconsistencies: {inconsistency_count}")

print("\n✅ Validation process completed.")
```

```python
with open('medicines.json', 'w') as f:
    json.dump(medicines_data, f, indent=4)

print("File 'medicines.json' has been created in your current directory.")
```
